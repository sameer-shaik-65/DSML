```python
# we didnt use outer becz in directors we have 2399 ids but in movies 
#we have 199 so outer wwill give manyy nan valesu for directors id we dont have data in ovies
data=movies.merge(directors,left_on="director_id",right_on="id",how="left")
```

```python
# convert ton million
data["budget"]=(data['budget']/1000000).round(2)
data['revenue']=(data['revenue']/1000000).round(2)
```

```python
# create a new copy of data
df=data.copy(deep=True)
df.head(2)
```

```python
# first movie and last movie and number of movies created  y each director
df_agg=df.groupby(["director_name"])[["year","title"]].aggregate({"year":["min","max"],"title":["count"]})
```

```python
def sub_avg(x):
  x=x-x.mean()
  return x
df.groupby(["director_name"])['budget'].transform(sub_avg)
```

```python
# create a new colm risky
def func(x):
  x["risky"]=(x["budget"]-x["revenue"].mean()) >=0
  return x
df.groupby(['director_name']).apply(func)
```
