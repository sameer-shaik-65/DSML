```python
# merge
users=pd.DataFrame([[1,"Shreyansh"],[2,"Rajat"],[3,"Sherlock"]],columns= ["userid","name"])
msgs=pd.DataFrame({"userid":[1,1,2],"msg":["hello","goodbye","tata"]})
```

```python
users
```

```python
users.merge(msgs,on="userid")
```

```python
# if we want ro do join on two columns
users.merge(msgs,left_on="userid",right_on="userid",how="inner")
```
